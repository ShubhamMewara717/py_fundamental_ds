Operators

```python
#Arthmetic operators
print(5+5)

print(5-5)

print(5*5)

print(5/2)

print(5//2)

print(5%2)

print(5**2)
```

```python
#Relationship operators
print(4>5)

print(4<5)

print(4>=4)

print(4<=4)

print(4==4)

print(4!=4)
```

```python
#Logistic 

print( 0 and 1)

print(0 or 1)

print(not 0)
```

```python
#Bitwise operators
 

#bitwise and
print( 2 & 3)


#bitwise or
print(2 | 3)

#bitwise xor
print(2 ^ 3)


print(4 >> 2)
print(4 << 2)
print(~2)
```

```python
#Assignment

a = 2
a+= 4
print(a)
a+=9
print(a)
a+= 8
print(a)

a%=2
print(a)
```

```python
#Membership operators in / not in

print("d" in"delhi")
print('s' not in 'shubham')
```

```python
first_num = int(input("entr the first number"))
second_num = int(input("entr the second number"))

op = input("enter the operation")

if op == "+":
    print(first_num + second_num)
elif op == "-":
    print(first_num - second_num)
elif op == "*":
    print(first_num * second_num)
elif op == "%":
    print(first_num % second_num)
else:
    print(first_num / second_num)
    
```
